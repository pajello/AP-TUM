```python
filenames=['ZUS.xlsx','Mappe1.xlsx']
sheetname1=['A1_25.2grad','A1_39.9grad','A1_49.6grad']
sheetname2=['A1_32.6grad','A1_42.2grad','A1_54grad']
```

```python
color_list=['red', 'blue','green', 'brown', 'black','orange']
label_list=['25.2°','39,9°','49,6°','32,6°','42,2°','54°']
#label_list_extra=['Transition points','Liquid gas coexistence boundary','critical point']
def ex4(filename,sheetname):
    df=pd.read_excel(filename,sheet_name=sheetname)
    #creating uarray for the volume
    VTemp=np.array(df['Volume[ml]']/(mol*1000)) #molar volume in liters
    n=len(VTemp)
    VErTemp=[0.1/(mol*1000)]*n

    V=unp.uarray(VTemp,VErTemp)

    #creating uarray for the pressure

    pTemp=np.array(df['p[atm]'])
    pErTemp=[0.5]*n

    p=unp.uarray(pTemp,pErTemp)
    
    #to match graph colors
    col=random.choice(color_list)
    i=color_list.index(col) #getting the index of the color we just choose
    
    
    poly=np.polyfit(VTemp,pTemp,6) #polinomial fitting 
    
    xx=np.linspace(0.15,1.55,100) #define x to plot 
    yy =np.polyval(poly, xx) #evaluate fit with xx 
    
    plt.plot(xx,yy,color=col,label=label_list[i])
    
    color_list.remove(col)
    label_list.remove(label_list[i])
    
   
    
    plt.errorbar(unp.nominal_values(V),unp.nominal_values(p),unp.std_devs(p),unp.std_devs(V),fmt=',',elinewidth=0.8,color=col)
    plt.xlabel('V [L/mol]')
    plt.ylabel('p [atm]')
    plt.ylim(10,50)
    
for name in sheetname1:
    ex4('ZUS.xlsx',name)
for name in sheetname2:
    ex4('Mappe1.xlsx',name)

#drawing phase change line    
critx=[0.195 ,0.23, 0.247 ,0.27,0.46,0.49,0.51,0.77]
crity=[22.8 ,29, 32.8 ,35.3,33.15,31,28.6,21.6]

#these arrays have exctra points to the one above, just to bias the line a bit towards the right
critx_cp=[0.195 ,0.23, 0.247 ,0.27,0.46,0.49,0.51, 0.62,0.77,1.4,1.5]
crity_cp=[22.8 ,29, 32.8 ,35.3,33.15,31,28.6, 24 ,21.6,16,15]
plt.scatter(critx,crity,color='purple',s=100,label='Transition points')
crit_param=np.polyfit(critx_cp,crity_cp,deg=5)
x=np.linspace(0.18,0.8,100)
y=np.polyval(crit_param,x)
plt.plot(x,y,'--',color='cornflowerblue',label='Liquid gas coexistence boundary')

#adding the top left portion to the dashed line
x_top=[0.17,0.2,0.3365656565656566,0.47313131313131324,0.50313131313131324]
y_top=[48,43,37.761510150579795,43,48]
crit_param_top=np.polyfit(x_top,y_top,deg=4)
x_top_cont=np.linspace(0.17,0.3365656565656566,100)
y_top_cont=np.polyval(crit_param_top,x_top_cont)
#plt.scatter(x_top,y_top)
plt.plot(x_top_cont,y_top_cont,'--',color='cornflowerblue')


#finding critical point, max of line
crit_point_y=max(y)
crit_point_x=float(x[np.where(y==crit_point_y)])
plt.scatter(crit_point_x,crit_point_y,s=100,marker='x',label='critical point')
print('The critical point is at coordinates',crit_point_x,' ',crit_point_y)
plt.legend(loc='upper right',prop={'size': 22})



plt.savefig('ex4.jpg', dpi=1000,bbox_inches="tight")
```
