```python
def plotting(x, y, prom=60, dist=20):
    '''
    Takes a noisy function, filters it, plots it unfiltered and filtered and finds the peaks.
    Parameters for the peaks:
    dist: horizontal distance between peaks 
    prom: prominence of peak'''
    y_fit = savgol_filter(y, 51, 10) # window size 51, polynomial order 5
    plt.plot(x,y,lw=1)
    plt.plot(x,y_fit,color='red',lw=0.7)
    (peak_pos, properties) = find_peaks(y_fit,distance=dist,prominence=prom, width=1)
    y_peak_list=[]
    for pos in peak_pos:
        y_peak=y_fit[pos]
        y_peak_list.append(y_peak)
    plt.scatter(peak_pos,y_peak_list,color='Green')
    return peak_pos, properties
def plotting2(x, y, prom=60, dist=20):
    '''
    Takes a noisy function, filters it, plots it unfiltered and filtered and finds the peaks.
    Parameters for the peaks:
    dist: horizontal distance between peaks 
    prom: prominence of peak'''
    y_fit = savgol_filter(y, 51, 10) # window size 51, polynomial order 5
    plt.plot(x,y,lw=1)
    plt.plot(x,y_fit,color='red',lw=0.7)
    (peak_pos, properties) = find_peaks(y_fit,distance=dist,prominence=prom, width=1)
    y_peak_list=[]
    for pos in peak_pos:
        y_peak=y_fit[pos]
        y_peak_list.append(y_peak)
    plt.scatter(peak_pos,y_peak_list,color='Green')
    return peak_pos, properties, y_fit
```

```python
peak_pos1, properties1 = plotting(x,y_list[0]) #Kobalt 60
peak_pos1 = peak_pos1[1:]
properties1['widths'] = properties1['widths'][1:]
energy1= np.array([1173,1333])
```

```python
peak_pos2, properties2 = plotting(x,y_list[1],1500) #Cesium 137
energy2= np.array([662])
```

```python
peak_pos3, properties3 = plotting(x,y_list[2],250,dist=185) #Natrium 22
energy3 = np.array([511,1275, 1787])
```

```python
peak_pos_tot = np.concatenate((peak_pos1,peak_pos2,peak_pos3))
energy_tot = np.concatenate((energy1,energy2,energy3))
halfwidth = np.concatenate((properties1['widths'],properties2['widths'],properties3['widths']))/2
halfwidth
```

```python
plt.scatter(peak_pos_tot,energy_tot)
```

```python
def linear_func(B, x):
    return B[0]*x+B[1]

x_data=peak_pos_tot
y_data=energy_tot
x_err=halfwidth

linear=Model(linear_func)
data=RealData(x_data, y_data, sx=x_err)
odr=ODR(data, linear, beta0=[-0.01, 0.0003])
out=odr.run()
out.pprint()
```

```python
def fit_function(channel,uncertainty):
    maxval = (2.85063295+0.03579405)*(channel+uncertainty)-34.90204376+13.47264313
    minval = (2.85063295-0.03579405)*(channel-uncertainty)-34.90204376-13.47264313
    return minval, maxval
```

```python
plt.errorbar(peak_pos_tot , energy_tot, yerr=None, xerr=halfwidth ,fmt='o', c='r')
x_v_lp=np.linspace(0,700,1000)
y_fit=linearFunc(x_v_lp,-34.90204376,2.85063295)
plt.plot(x_v_lp,y_fit)
plt.xlabel('Channel')
plt.ylabel('Energy [keV]')
plt.title('Calibration curve fit')
plt.savefig('Calibration curve fit.png',bbox_inches="tight")
```

```python
uncertainty_thorium = np.round(properties9['widths'][1:]/2).astype(np.int64) # half width rounded to integer
```

```python
peak_pos9
```

```python
def energy_error(peakpos,err):
    energies_peaks=[]
    errors=[]
    for i in range(len(peakpos)):
        eV = fit_function(peakpos[i],err[i])
        eV_v=np.sum(eV)/2 #finding energy level with uncertainty
        eV_u=abs(eV[0]-eV[1])/2
        energies_peaks.append(eV_v)
        errors.append(eV_u)
    return energies_peaks, errors
```

```python
energies_thorium, errors_thorium = energy_error(peak_pos9, uncertainty_thorium)
```

```python
f = open("Thorium.txt", "a")
text_list = [f'Die peaks sind bei {peak_pos9}\n',
            f'Mit dem fit errechen wir fuer den linken peak energien von {energies_thorium}\n',
            f'mit Fehlern: {errors_thorium}\n',
            f'Fehler von fitkurve und halfwidth miteingerechnet\n',
            f'das layout ist anders hier weil es zu viele lineen waren, musste mir eine funktion schreiben',]
f.writelines(text_list)
f.close()
```

```python
def plotting3(x, y, prom=60, dist=20):
    '''
    Takes a dataset, plots it unfiltered and filtered and finds the peaks.
    Parameters for the peaks:
    dist: horizontal distance between peaks 
    prom: prominence of peak'''
    plt.plot(x,y,lw=1)
    (peak_pos, properties) = find_peaks(y,distance=dist,prominence=prom, width=1)
    y_peak_list=[]
    for pos in peak_pos:
        y_peak=y[pos]
        y_peak_list.append(y_peak)
    plt.scatter(peak_pos,y_peak_list,color='Green')
    return peak_pos, properties
```

```python
C60=np.array(df2['Kalibrierung Co60'],dtype='float')
C60=C60[:100]
Unter=np.array(df2['Untergrund Messung'],dtype='float')
x2=np.array(df2['Kanal'], dtype='float')
x2= x2[:100]
```

```python
#plt.scatter(peaksC60,energyC60)
def linear_func(B, x):
    return B[0]*x+B[1]

x_dataC60=peaksC60
y_dataC60=energyC60
x_errC60=halfwidthC60

linear=Model(linear_func)
data=RealData(x_dataC60, y_dataC60, sx=x_errC60)
odr=ODR(data, linear, beta0=[-0.01, 0.0003])
out=odr.run()
out.pprint()
```

```python
def fit_functionC60(channel):
    maxval = (48.97175182+0.2308623)*channel-8.53994186+10.3826063
    minval = (48.97175182-0.2308623)*channel-8.53994186-10.3826063
    return (minval+ maxval)/2,abs(minval-maxval)/2
```

```python
def fit_func_err(channel):
    m=ufloat(49.09694786,0.5027589)
    b=ufloat(3.32119543,16.63479003)
    return m*channel+b
```

```python
E,u_E=fit_functionC60(1000) #energie in keV
print(f'{E/1000} MeV pm {u_E/1000}')
```
